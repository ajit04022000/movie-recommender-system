```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.head()
```

```python
credits.head()
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.head(2)
```

```python
movies.columns
```

```python
# removing some unusefull columns based on recomdetaion type model and keep the usefull columns
# genres , id , keywords , title , overview ,release_date, cast ,crew
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
# the columns overwiew is in string format so i need to converte this into a list format
movies['overview'][0]
```

```python
movies['overview']=movies['overview'].apply(lambda x: x.split())
```

```python
movies['genres'][0]
```

```python
# ["Action","Adventure","Fantasy", "Science Fiction"] we need to find this from 'genres' column so we use ast.literal_eval 
# function from ast librabary
import ast
```

```python
def convert(obj):
    L =[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
## hear we are finding the top 3 actors of that specific fillms
def convert3(obj):
    L =[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
def fatch_diractor(obj):
    """finding in crew columns
    where job:Diractor"""
    L =[]
    for i in ast.literal_eval(obj):
        if i['job'] == "Director":
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fatch_diractor)
```

```python
movies.head()
```

```python
# now i am going to remove the space of all entites present in 'genres', 'keywords', 'cast', 'crew' to make as an one netites
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(' ','') for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(' ','') for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(' ','') for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
movies.head()
```

```python
## now i am making a new column as "tags" and concatinate these 'genres', 'keywords', 'cast', 'crew' to make as an one 
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] +movies['cast']+movies['crew'] 
```

```python
movies.head()
```

```python
# making a new data frame it contain movies_id , title, and tags . 
new_df = movies[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
# now i am going to convert the tags from list to string
new_df['tags'] = new_df['tags'].apply(lambda x : " ".join(x)) 
```

```python
new_df.head()
```

```python
# now i am going to convert the tags into lower case:
new_df['tags'] = new_df['tags'].apply(lambda x :x.lower()) 
```

```python
new_df.head()
```

```python
## now i am going to apply staming 
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
## now we are going to do the vectorization(NLP) thorugh bag of word:
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=500,stop_words="english")
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
## hear we can see that there are 4806 no of movies present it means we
## have 4806 no of vectors are also present and each vectors contain 5000 words numbers are present


## and now as we know if we are working on a highdimkeansional data the euclidan distance not a reliable measure to capture all these
## so now i am calucating the cosing distance (hear the thita value is low is denote the movies are more similar to us)
```

```python
## i am going to import librabaery to perform cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
## now i am going to make a function for recomendation of five movies as our input is
def recommend(movie):
    
    '''1st- i am finding the index of that specfic movies as input
    2nd- sorting the index i got from 1st step
    3rd- than fatching the top 5 movies and give in out put'''
    
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list  = sorted(list(enumerate(distances)),reverse = True,key = lambda x: x[1])[1:6]
    
    for i in movies_list :
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
